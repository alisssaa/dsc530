**Analytical Distribution**: characterized by a CDF that is a mathematical function. Can be used to model empirical distributions.

**Exponential Distribution**<br>
Occurs when we look at a series of events and measure the time between events, the **interarrival time**. If events are equally likely to occur at anytime, it results in an exponential distribution.

```python
import pandas
import numpy as np

import analytic
import thinkplot
import thinkstats2
```

```python
from analytic import ReadBabyBoom

df = ReadBabyBoom()
diffs = df.minutes.diff()
cdf = thinkstats2.Cdf(diffs, label = "actual")

thinkplot.Cdf(cdf)
thinkplot.Show(xlabel = "minutes", ylabel = "CDF")
```

To test whether it fits an exponential distribution, we can graph it alongside the **complementary CDF**.

```python
thinkplot.Cdf(cdf, complement = True)
thinkplot.Show(xlabel = "minutes",
               ylabel = "CCDF",
               yscale = "log")
```

```python
def MakeNormalPlots(weights):
    mean = weights.mean()
    std = weights.std()
    
    xs = [-4, 4]
    fxs, fys = thinkstats2.FitLine(xs, inter = mean, slope = std)
    thinkplot.Plot(fxs, fys, color = "gray", label = "model")
    
    xs, ys = thinkstats2.NormalProbability(weights)
    thinkplot.Plot(xs, ys, label = "birth weights")
```

**Kernel Density Estimate**: takes a sample and finds an appropriately smooth PDF that fits the data

```python
import random
import math

mean = 60
std = 7

sample = [random.gauss(mean, std) for i in range(500)]
sample_pdf = thinkstats2.EstimatedPdf(sample)
thinkplot.Pdf(sample_pdf, label = 'sample KDE')
```
