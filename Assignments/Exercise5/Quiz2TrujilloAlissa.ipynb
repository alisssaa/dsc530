```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = pd.read_csv("Quiz 2_data.csv")
```

```python
print("Rows     :",data.shape[0])
print("Columns  :",data.shape[1])
print("\nFeatures :\n",data.columns.tolist())
print("\nMissing Values :",data.isnull().sum().sum())
print("\nColumns with Missing Values:\n",data.isnull().any())
```

```python
data.drop(["id", "name", "host_name", "last_review"], axis = 1, inplace = True)
print(data.head())
```

```python
data.reviews_per_month.fillna(0, inplace = True)
print(data.isnull().any())
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize = (13, 7))
plt.title("Neighborhood Group")
plt.pie(data.neighbourhood_group.value_counts(), 
        labels = data.neighbourhood_group.value_counts().index, 
        autopct = '%1.1f%%', startangle = 180)
plt.show()
```

```python
plt.figure(figsize = (13, 7))
plt.title("Map of Neighborhood Group")
sns.scatterplot(data.longitude, data.latitude, hue = data.neighbourhood_group)
plt.ioff()
plt.show()
```

```python
plt.figure(figsize = (13, 7))
plt.title("Type of Room")
sns.countplot(data.room_type, palette = "muted")
fig = plt.gcf()
plt.show()
```

```python
plt.figure(figsize = (13, 7))
plt.title("Rooom Type on Neighborhood Group")
sns.countplot(data.neighbourhood_group, hue = data.room_type,
             palette = "muted")
plt.show()
```
